```python
#Cargar librerias
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import pandas as pd # cargar conjuntos de datos externos
import sklearn.linear_model # contiene diferentes modelos de ML
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
calificaciones = np.clip(calificaciones, 0, 100)  # Ajustar las calificaciones a estar en el rango [0, 100]
```

```python
#Procesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
print("Forma de train_data:", len(train_data))
print("Forma de train_labels:", len(train_labels))
print("Forma de test_data:", len(test_data))
print("Forma de test_labels:", len(test_labels))
```

```python
#Visualizar datos
def visualize_data(horas_estudio, calificaciones):
    # Crear DataFrame
    df = pd.DataFrame({'Horas de Estudio': horas_estudio.flatten(), 'Calificaciones': calificaciones.flatten()})
    # Visualizar datos en tabla
    print("Datos Generados:")
    print(df.head(10))  # Mostrar las primeras 10 filas
    print("\n")
    # Visualizar datos en diagrama de dispersión
    plt.scatter(horas_estudio, calificaciones, color='blue')
    plt.title('Datos Generados')
    plt.xlabel('Horas de estudio')
    plt.ylabel('Calificaciones')
    plt.show()

visualize_data(horas_estudio, calificaciones)

```

```python
# Seleccionar el modelo a entrenar: Regresión lineal

model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(train_data, train_labels)
```

```python
# Predicciones con el modelo entrenado
X_new= [[9]]  # Horas de estudio
print("Predicción 1:", model.predict(X_new))
```

```python
#Predicción 2
X_new2 = [[5]]  # Horas de estudio
print("Predicción 2:", model.predict(X_new2))
```

```python
# Calculo del error cuadrático medio
predicciones = model.predict(test_data)
error = mean_squared_error(test_labels, predicciones)
print("Error cuadrático medio:", error)
```

```python
# Visualización final con el modelo entrenado
plt.scatter(train_data, train_labels, color='red')
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.title("Modelo de Regresión Lineal")
plt.ylim(0, 25)  # Ajustar el límite del eje y
plt.plot(horas_estudio, model.predict(horas_estudio), color='blue')
plt.show()
```
